```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.image as img
```

```python
bottom_left=cv2.imread('ImageCollage dataset (1)/bottom_left.jpg')
bottom_right=cv2.imread('ImageCollage dataset (1)/bottom_right.jpg')
top_left=cv2.imread('ImageCollage dataset (1)/top_left.jpg')
top_right=cv2.imread('ImageCollage dataset (1)/top_right.jpg')
center=cv2.imread('ImageCollage dataset (1)/center.jpeg')
```

```python
center.shape
```

```python
plt.imshow(center)
```

```python
bottom_left1=cv2.cvtColor(bottom_left,cv2.COLOR_BGR2RGB)
bottom_right1=cv2.cvtColor(bottom_right,cv2.COLOR_BGR2RGB)
top_left1=cv2.cvtColor(top_left,cv2.COLOR_BGR2RGB)
top_right1=cv2.cvtColor(top_right,cv2.COLOR_BGR2RGB)
center1=cv2.cvtColor(center,cv2.COLOR_BGR2RGB)
```

```python
resize_image_bottom_left1=cv2.resize(bottom_left1,(200,200))
resize_image_bottom_right1=cv2.resize(bottom_right1,(200,200))
resize_image_top_left1=cv2.resize(top_left1,(200,200))
resize_image_top_right1=cv2.resize(top_right1,(200,200))
resize_image_center1=cv2.resize(center1,(100,100))
```

```python
resize_image_bottom_left1=cv2.copyMakeBorder(resize_image_bottom_left1, top=5, bottom=10, left=10, right=5, borderType=cv2.BORDER_CONSTANT)
resize_image_bottom_right1=cv2.copyMakeBorder(resize_image_bottom_right1, top=5, bottom=10, left=5, right=10, borderType=cv2.BORDER_CONSTANT)
resize_image_top_left1=cv2.copyMakeBorder(resize_image_top_left1, top=10, bottom=5, left=10, right=5, borderType=cv2.BORDER_CONSTANT)
resize_image_top_right1=cv2.copyMakeBorder(resize_image_top_right1, top=10, bottom=5, left=5, right=10, borderType=cv2.BORDER_CONSTANT)
resize_image_center1=cv2.copyMakeBorder(resize_image_center1, top=10, bottom=10, left=10, right=10, borderType=cv2.BORDER_CONSTANT)
```

```python
plt.imshow(resize_image_center1)

```

```python
plt.imshow(resize_image_bottom_right1)
```

```python
Horizontal1=np.hstack([resize_image_top_left1,resize_image_top_right1])
Horizontal2=np.hstack([resize_image_bottom_left1,resize_image_bottom_right1])
```

```python
Vertical1=np.vstack([Horizontal1,Horizontal2])
```
